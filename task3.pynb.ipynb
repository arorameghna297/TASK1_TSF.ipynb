##LINE PLOT


```
data.groupby('Sub-Category')['Profit','Sales'].agg(sum).plot.line()
plt.title('Total Profit & Sales per Sub-Category');
```

```
data_state=data.groupby(['State'])[['Quantity','Sales','Discount','Profit']].mean()
data_state.head(10)
```


##PIE CHARTS

```
plt.figure(figsize=(15,15))
data.groupby("Sub-Category").sum()["Sales"].plot.pie(autopct="%1.1f%%")
plt.title('Sub_Category-wise analysis of Sales',fontsize=20)
```
